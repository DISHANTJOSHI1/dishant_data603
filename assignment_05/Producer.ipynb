# Producer

```python
import time 
import json
import pandas as pd
from kafka import KafkaProducer
from urllib.request import urlopen 
from time import sleep 

url = 'http://api.open-notify.org/iss-now.json'

def producer_ftn(message):
    return json.dumps(message).encode('utf-8')

producer = KafkaProducer(bootstrap_servers=['localhost:9092'], 
                                            value_serializer=producer_ftn, 
                                            retries = 5)
                         
if __name__ == '__main__':
    time_ends = time.time() + 60 * 60
    while time.time() < time_ends:
        response = urlopen(url)
        data_extracted = json.loads(response.read())
        print(data_extracted)
        message = data_extracted 
        producer.send ('messages', data_extracted) # producer sending messages in json format to consumer
        sleep(5) # 5 seconds timer
        producer.flush () #Synchronizing the producer
```
